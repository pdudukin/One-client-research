```python
def find_all_event_ids_by_users(list_of_users_ids):
    event_ids_list = []
    for user_id in list_of_users_ids:
        event_ids_list.extend(list(data[data.user==user_id]['registration_event_id'].values))
    return list(set(event_ids_list))
```

```python
def recommended_event_ids(user_event_ids_list, other_users_event_ids_list):
    return [x for x in other_users_event_ids_list if x not in user_event_ids_list]
```

```python
def give_recommendations(user_id):
    list_of_similar_users = []
    user_id_categories = list(set(data[data.user==user_id]['name'].values))
    print(user_id_categories)
    
    for other_user_id in dd.index:
        if user_id!=other_user_id and spatial.distance.cosine(dd[dd.index==user_id], dd[dd.index==other_user_id])<0.3:
            #print(user_id, other_user_id, 1 - spatial.distance.cosine(dd[dd.index==user_id], dd[dd.index==other_user_id]))
            list_of_similar_users.append(other_user_id)
    
    if len(list_of_similar_users)>0:
            print('User ID: {}.'.format(user_id))
            print('Recommended event ids: {}'.format(recommended_event_ids([user_id],find_all_event_ids_by_users(list_of_similar_users))))
    else:
        print('No similar users found for user id: {}'.format(user_id))
```
